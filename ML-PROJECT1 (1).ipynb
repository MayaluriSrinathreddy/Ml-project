```python
X=df.drop(columns=["ProdTaken"])
y=df["ProdTaken"]
```

```python
# segregate the columns train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
df.info()
```

```python
# separate categorical and numerical
X_train_cat=X_train.select_dtypes(include=["object"])
X_train_num=X_train.select_dtypes(include=["int64","float64"])
X_train_cat.shape
```

```python
# resscaling numerical features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_num_transformed = pd.DataFrame(scaler.fit_transform(X_train_num), columns=scaler.get_feature_names_out(), index=X_train_num.index)
```

```python
# onehotencoding the categorical features
from sklearn.preprocessing import OneHotEncoder
one_hot=OneHotEncoder(drop="first",sparse_output=False,handle_unknown="ignore")
X_train_cat_transformed=pd.DataFrame(one_hot.fit_transform(X_train_cat),columns=one_hot.get_feature_names_out(),index=X_train_cat.index)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
logistic.fit(X_train_transformed,y_train)
```

```python
y_test_ppredict=logistic.predict(X_test_transformed)
```

```python
y_test_ppredict
```

```python
from sklearn import metrics
log_predict=metrics.accuracy_score(y_test_ppredict,y_test)
log_predict
```

```python
from sklearn.neighbors import KNeighborsClassifier
neighbors=KNeighborsClassifier()
neighbors.fit(X_train_transformed,y_train)
```

```python
y_test_predic=neighbors.predict(X_test_transformed)
```

```python
knn_predict=metrics.accuracy_score(y_test_predic,y_test)
knn_predict
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(X_train_transformed,y_train)
```

```python
y_test_pre=tree.predict(X_test_transformed)
```

```python
dt_predict=metrics.accuracy_score(y_test_pre,y_test)
dt_predict
```

```python
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier()
forest.fit(X_train_transformed,y_train)
```

```python
y_test_pred=forest.predict(X_test_transformed)
```

```python
rf_predict=metrics.accuracy_score(y_test_pred,y_test)
rf_predict
```

```python

from sklearn.naive_bayes import GaussianNB
naive_baye=GaussianNB()
naive_baye.fit(X_train_transformed,y_train)
```

```python
y_test_predi=naive_baye.predict(X_test_transformed)
```

```python
%%time
nb_predict=metrics.accuracy_score(y_test_predi,y_test)
nb_predict
```

```python
Algorithms = ["Logistic Regression","knn","decission tree","Random Forest","navies baisd"]
```

```python
Accuracy = [log_predict,knn_predict,dt_predict,rf_predict,nb_predict]
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=Algorithms, y=Accuracy, palette="tab10")
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.title("Accuracy of Given Models")
for i, container in enumerate(ax.containers):
    ax.bar_label(container, fontsize=10)

plt.show()
```
